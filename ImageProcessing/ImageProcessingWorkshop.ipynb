## Making a sequence of frames

```python
arr = np.zeros((100, 200))
```

```python
# Let's make a fountain over time
# For that we need the projectile motion equations
def projectile_location(time, angle, v0):
    j = v0 * time * np.cos(angle)
    i = v0 * time * np.sin(angle) - 0.5 * 9.81 * time ** 2
    return i, j
# Let's test this
angle = np.pi / 4.0
v0 = 4
time = np.linspace(0, 1) * 2 * v0 * np.sin(angle) / 9.81
y, x = projectile_location(time, angle, v0)
plot(x, y)
xlabel("x distance")
ylabel("y distance")
axis('scaled')
axis([0, x.max(), 0, y.max() * 1.1])
np.sin(angle)
```

```python
# Let's make 100 images
images = []

# Let's keep track of fountain particles
max_number_particles = 10
avg_initial_velocity = 10.0
std_initial_velocity = 1.0
avg_angle = np.pi / 2.0
std_angle = np.pi / 30.0
time_scale = 2.0 * avg_initial_velocity / 9.81 / 100.0 * 2  # 10 cycles of particles hitting ground
IJ_scale = 10
time = []
angle = []
v0 = []
for i in xrange(100):
    arr = np.zeros((100, 200, 3), dtype=np.uint8)
    # Initialize a random number of particles
    n_new_part = int(np.round(np.random.rand(1) * max_number_particles))
    time += [i] * n_new_part
    angle += (np.random.randn(n_new_part) * std_angle + avg_angle).tolist()
    v0 += (np.random.randn(n_new_part) * std_initial_velocity + avg_initial_velocity).tolist()
    
    # Make lists into arrays
    time_arr = (i - np.array(time)) * time_scale
    angle_arr = np.array(angle)
    v0_arr = np.array(v0)
    
    # find particle locations
    II, JJ = projectile_location(time_arr, angle_arr, v0_arr)
    I = ((arr.shape[0] - 1 - II * IJ_scale)).astype(int)
    J = ((JJ * IJ_scale + arr.shape[1] / 2)).astype(int)
    # Find particules that are not in domain anymore
    ids = (I < 0) | (J < 0) | (I >= arr.shape[0])  | (J >= arr.shape[1])
    I = I[~ids]
    J = J[~ids]
    
    # rebuild lists, remove the above elements
    time = [t for ii, t in zip(ids, time) if not ii]
    angle = [a for ii, a in zip(ids, angle) if not ii]
    v0 = [v for ii, v in zip(ids, v0) if not ii]
    
    # Finally, draw these points on the image
    arr[I, J, :] = 255
    
    # And save the image
    img = PIL.Image.fromarray(arr)
    img.save('anim/%05d.png' % i)
```

Well, that doesn't look very exciting, or fountainy. Is there a way for me to modify these images so that it looks at least more like a fountain? 

```python
# Step 1: Get a list of the images
import os  # This standard python package helps with operating system related task, like getting a list of files
files = os.listdir('anim')
files.sort()
files[:3]
```

```python
# Let's import the image filter module
import PIL.ImageFilter
# Let's experiment on the last image
image_blurred = image.filter(PIL.ImageFilter.GaussianBlur).filter(PIL.ImageFilter.GaussianBlur).filter(PIL.ImageFilter.SMOOTH)
imshow(image_blurred, interpolation='none')
```

```python
# Now it's a little dark, let's also make it bluer
image_blurred_arr = np.array(image_blurred)
image_blurred_arr[:, :, 0] *= 0.75
image_blurred_arr[:, :, 1] *= 1.5
image_blurred_arr[:, :, 2] *= 2.5
image_blurred_arr[image_blurred_arr > 255] = 255
imshow(image_blurred_arr, interpolation='none')
image_blurred_arr.max()
```

```python
# Okay, let's do that for all the images
os.mkdir('anim_blurred')
for fil in files:
    image = PIL.Image.open(os.path.join("anim", fil))
    image_blurred = image.filter(PIL.ImageFilter.GaussianBlur).filter(PIL.ImageFilter.GaussianBlur).filter(PIL.ImageFilter.SMOOTH)
    image_blurred_arr = np.array(image_blurred)
    image_blurred_arr[:, :, 0] *= 0.75
    image_blurred_arr[:, :, 1] *= 1.5
    image_blurred_arr[:, :, 2] *= 2.5
    image_blurred_arr[image_blurred_arr > 255] = 255
    image = PIL.Image.fromarray(image_blurred_arr)
    image.save(os.path.join('anim_blurred', fil))
```

So, we used python and PIL to blur a folder full of images, but we could also cropped, resized, or watermarked all these images just as easily.

```python
# And for fun, let's look at the images we've created using an interactive widget
from IPython.html.widgets import interactive
# We have to define a function that interactive will write
# Load all the pictures into memory
movie_files = np.zeros((100, ) + image_blurred_arr.shape, dtype=np.uint8)
for i in xrange(100):
    image = PIL.Image.open(os.path.join("anim_blurred", '%05d.png' % i))
    movie_files[i, :, :, :] = np.array(image)
def show_image(i):    
    imshow(movie_files[i, :, :, :], interpolation='none')
    axis('off')
    show()
widget = interactive(show_image, i=(0, 99))
ii,  = widget.children
```
